```python
import numpy as np
import matplotlib.pyplot as plt
```

$sin(x) = x  - \frac {x^3} {3!} + \frac{x^5} {5!}$

```python
def term(x, n):
    sign = 1 if n%2 == 0 else -1
    order = 2*n + 1
    numerator = x**order
    denominater = np.prod(np.arange(1,order+1,1))
    return sign*numerator/denominater
```

```python
x = np.linspace(-10*np.pi,10*np.pi,1000)

y_real = np.sin(x)

add = x.copy()
y_approx = add.copy()

for i in range(40):
    add = -add * x**2 / ((2*i+2) * (2*i+3))
    y_approx += add
    print(add)
```

```python
fig = plt.figure(dpi=300,figsize=(8,4))
ax = fig.add_subplot(111)
ax.plot(x,y_real,label="real")
ax.plot(x,y_approx,label="approxmation")
ax.set_ylim((-1.5,1.5))
ax.legend()
plt.show()
```
