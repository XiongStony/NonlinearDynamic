```python
from scipy.integrate import odeint
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(12,12))
```

```python
omega = 2
omegasq = omega**2
a, b, c, d = 0, 1, omegasq, -.5
```

```python
def dx_dt(xx,t):
    x = xx[0]
    y = xx[1]
    term1 = a*x + b*y
    term2 = -omegasq*np.sin(x) + d*y
    return [term1, term2]
```

```python
maxtime = 20
xmin = -8
xmax = 8
ts = np.linspace(0,maxtime,400)
ic = np.linspace(xmin,xmax,8)
```

```python
for r in ic:
    for s in ic:
        x0 = [r,s]
        xs = odeint(dx_dt, x0, ts)
        plt.plot(xs[:,0],xs[:,1],"b")

plt.xlabel('x')
plt.ylabel('y')
plt.xlim(xmin,xmax)
plt.ylim(xmin,xmax)
```
